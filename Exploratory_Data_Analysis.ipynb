Simplified function to convert string numbers to integers, non-numeric to 0

```python
def convert_to_int(str_values):
    try:
        return int(str_values)
    except ValueError:
        return float(str_values)
```

Apply the conversion function to all features that were supposed to be numerical features.

```python
df['text_reviews_count'] = df['text_reviews_count'].apply(convert_to_int)
df['num_pages'] = df['num_pages'].apply(convert_to_int)
df['average_rating'] = df['average_rating'].apply(convert_to_int)
```
